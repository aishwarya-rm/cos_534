```python
import pickle
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

palette = sn.color_palette('hls', 12)
```

```python
kidney_results = pickle.load(open('kidney_results.pkl', 'rb'))
ventilation_results = pickle.load(open('ventilation_results.pkl', 'rb'))
```

```python
overall_data = []
for m in kidney_results.values():
    overall_data.append(['accuracy', m['overall_accuracy']])
    overall_data.append(['precision', m['overall_precision']])
    overall_data.append(['recall', m['overall_recall']])
overall_df = pd.DataFrame(overall_data, columns=['measurement', 'value'])
```

```python
sn.set_style('darkgrid')
sn.set(font_scale=2)
plt.figure(figsize=(12,8))
ax = sn.boxplot(x="value", y="measurement", hue="measurement",
              data=overall_df, palette='Set3')
ax.set_title('[Kidney Transplant] Overall Metrics in Prediction Task')
ax.get_figure().savefig('kidney_overall_metrics.png', bbox_inches = "tight")
```

```python
subgroup_data = []
columns = ['ethnicity', 'gender', 'subgroup', 'accuracy', 'ppv', 'tpr', 'fpr', 'fnr', 'ba']
for m in kidney_results.values():
    for d in zip(m['group_list'], m['accuracy_list'], m['ppv_list'], m['tpr_list'], m['fpr_list'], m['fnr_list'], m['ba_list']):
        e = d[0][:-2]
        if e == 'AMERICAN INDIAN/ALASKA NATIVE':
            e = 'AM. INDIAN/ALASKA NATIVE'
        g = d[0][-1]
        if g == '0':
            g = 'Male'
        else:
            g = 'Female'
        
        subgroup = d[0]
        subgroup_data.append([e, g, subgroup, d[1], d[2], d[3], d[4], d[5], d[6]])
```

```python
subgroup_df = pd.DataFrame(subgroup_data, columns=columns)
```

```python
sn.set_style('darkgrid')
plt.figure(figsize=(12,8))
# plt.xticks(rotation=30)
ax = sn.boxplot(x="accuracy", y="ethnicity", hue="gender",
              data=subgroup_df, palette='Set3')
ax.set_title('[Kidney Transplant] Accuracy by Ethnicity and Gender')
ax.get_figure().savefig('kidney_accuracy.png', bbox_inches = "tight")
plt.figure(figsize=(12,8))
# plt.xticks(rotation=30)
ax = sn.boxplot(x="tpr", y="ethnicity", hue="gender",
              data=subgroup_df, palette='Set3')
ax.set_title('[Kidney Transplant] TPR by Ethnicity and Gender')
ax.get_figure().savefig('kidney_tpr.png', bbox_inches = "tight")
plt.figure(figsize=(12,8))
# plt.xticks(rotation=30)
ax = sn.boxplot(x="fpr", y="ethnicity", hue="gender",
              data=subgroup_df, palette='Set3')
ax.set_title('[Kidney Transplant] FPR by Ethnicity and Gender')

ax.get_figure().savefig('kidney_fpr.png', bbox_inches = "tight")
plt.figure(figsize=(12,8))
# plt.xticks(rotation=30)
ax = sn.boxplot(x="ba", y="ethnicity", hue="gender",
              data=subgroup_df, palette='Set3')
ax.set_title('[Kidney Transplant] Bias Amp. by Ethnicity and Gender')
ax.get_figure().savefig('kidney_ba.png', bbox_inches = "tight")
```
